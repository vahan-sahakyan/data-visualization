```python
import numpy as np
import matplotlib.pyplot as plt

import matplotlib as mpl
```

```python
X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
```

```python
fig, ax = plt.subplots()
ax.plot(X, Y)
plt.show()
```

```python
X = np.random.uniform(0,1,100)
Y = np.random.uniform(0,1,100)
fig, ax = plt.subplots()
ax.scatter(X, Y)
```

```python
X = np.arange(10)
Y = np.random.uniform(1,10,10)
fig, ax = plt.subplots()
ax.bar(X, Y)
```

```python
Z = np.random.uniform(0,1,(12,12))
fig, ax = plt.subplots()
ax.imshow(Z)
ax.invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.contourf(Z)
```

```python
Z = np.random.uniform(0,1,4);
fig, ax = plt.subplots();
ax.pie(Z);
```

```python
Z = np.random.uniform(0,1,100);
fig, ax = plt.subplots();
ax.hist(Z);
```

```python
X = np.arange(5);
Y = np.random.uniform(0,1,5);
fig, ax = plt.subplots();
ax.errorbar(X,Y,Y/4);
```

```python
Z = np.random.normal(0,1,(100,4));
fig, ax = plt.subplots();
ax.boxplot(Z);
```
