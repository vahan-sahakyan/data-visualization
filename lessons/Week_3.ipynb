### Seaborn plots



    

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Scatter plot

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'total_bill', 
                y = 'tip', 
                data = tips,
                hue = 'sex',
               size = 'size'
               )
```

```python
sns.relplot(x = 'total_bill', 
                y = 'tip', 
                data = tips,
                hue = 'sex',
               size = 'size',
             kind='scatter')
```

```python
sns.jointplot(x = 'total_bill', 
                y = 'tip', 
                data = tips,
              #  hue = 'sex',
             kind='kde'
             )
```

### Line plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize = (10,6))
sns.lineplot(x = 'year', 
             y = 'passengers', 
             data = flights,
            #  estimator = np.median, 
            #  marker = 'o',
            #  palette="plasma",
             hue = 'month',                        
             style="month"

            )
```

```python
sns.relplot(x = 'year', 
             y = 'passengers', 
             data = flights, 
             
            kind='line',
           hue = 'month')
```

### Distribution plots
* Histogram
* KDE
* ecdf


```python
plt.figure(figsize=(10,6))
sns.histplot(data=tips,
             x="total_bill",
            stat="count",
            hue = 'sex',
            kde=True,            
            bins=25,
            multiple = 'stack'
            )
```

```python
#plt.figure(figsize=(10,6))
sns.displot(data=tips,
             x="total_bill", 
            # kind = 'kde', # or kde
            hue = 'sex',
            kde=True, # or kind         
            bins=25,
            element="step",
            # multiple = 'dodge' #'dodge'

           row = 'time',
           col = 'day'
            )
```

```python
# pmf vs cdf

# Probability mass function
# Cumulative distribution function

plt.figure(figsize=(12,6))
sns.ecdfplot(data=tips, 
             x="total_bill"         
            )
sns.rugplot(data=tips, 
             x="total_bill"         
            )
```

```python
diamonds = sns.load_dataset('diamonds')
print(diamonds.info())
print('-'*80)
print(diamonds.head())
```

```python
sns.countplot(x = 'color', data =diamonds, hue="color")
```

```python
sns.countplot(x = 'color', data =diamonds, 
              order = ['G', 'E', 'F', 'H', 'D', 'I', 'J'],
              hue = 'cut', 
              # hue_order = ['Ideal', 'Premium',],
              lw = 4, ec = 'k', hatch = '/'

    )
```

```python
print('\ncolor:')
print(diamonds.color.value_counts().index)
print('\ncut:')
print(diamonds.cut.value_counts().index)
```

```python
sns.boxplot(data=tips,
            y = "total_bill" ,
            x = 'day',
            hue = 'sex'
            
            
             )
```

```python
plt.figure(figsize=(14, 5))
sns.stripplot(data=tips,
              x="day",
              y="tip",
              hue="sex",
              dodge=True,
              palette = "muted"
             )
```

```python
plt.figure(figsize=(14, 5))
sns.swarmplot(data=tips,
              x="day",
              y="tip",
              hue="sex",
              dodge=True,
              palette = "muted"
             )
```

```python
plt.figure(figsize=(14, 5))
sns.violinplot(data=tips,
              x="day",
              y="tip",
              hue="sex", 
              split = True
             )
```

```python
plt.figure(figsize=(14, 5))
sns.pointplot(data=tips,
              x="day",
              y="tip",
              hue="sex", 
              dodge = True, 
              errorbar=('ci', 9) # ci = confidence interval
             )
```

```python
plt.figure(figsize=(14, 5))
sns.pointplot(data=tips,
              x="day",
              y="tip",
              hue="sex", 
              dodge = True, 
              errorbar="sd" # sd = standard deviation
             )
```

```python
plt.figure(figsize=(14, 5))
sns.pointplot(data=tips,
              x="day",
              y="total_bill",
              hue="sex", 
              dodge = True, 
              errorbar="se" # se = standard error
             )

# zero hypothesis in simple words is that there is no difference in total bill between men and women
# we can reject the null hypothesis if p-value < 0.05
# we can not reject the null hypothesis if p-value > 0.05

```
