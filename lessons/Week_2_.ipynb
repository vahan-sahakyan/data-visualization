### pyplot interface (procedural, inheritance of matlab))

```python
x1 = np.arange(-2,8,0.5)
y1 = x**2
x = np.arange(0,10,0.5)
y = x**2
x2 = np.arange(2,12,0.5)
y2 = x**2


plt.plot(x1, y1)
plt.plot(x, y)
plt.plot(x2, y2)
# plt.show()
```

```python
plt.plot(x,y,
         color='blue', #'b', '#1f77b4', (0,0,1)
         marker='o', 
         linestyle='dotted', # '--'
        )
```

```python
plt.plot(x,y,'o-.c'); # fmt = '[marker][line][color]'
```

```python
# Add title, lables, limits
plt.plot(x,y,'o--b',markeredgecolor = 'g',)
plt.title('$f(x)=x^2$',
         family = 'sans-serif',
         color = 'r',
         weight = "bold",
         size = 24
)
plt.xlabel('x')
plt.ylabel('f(x)');
plt.yticks(rotation=25, ha="right")
plt.xticks(rotation=90, ha="right")
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
```

```python
custom_size = 20

plt.figure(figsize=(8,5))
plt.plot(x1,y1,'o--y',markeredgecolor = 'y', markersize = custom_size,
         label = '$f(x)=x^2$')
plt.plot(x,y,'o--c',markeredgecolor = 'c', markersize = custom_size,
         label = '$f(x)=x^2$')
plt.plot(x2,y2,'o--r',markeredgecolor = 'r', markersize = custom_size,
         label = '$f(x)=x^2$')
plt.title('$f(x)=x^2$',
         family = 'serif',
         color = 'r',
         weight = 'bold',
         size = 16) 
plt.xlabel('x', fontdict = font)
plt.ylabel('f(x)', fontdict = font)
plt.legend(loc='lower right')
plt.grid()
#plt.figure(figsize=(8,5));
```

```python
# 'b' = blue.
# 'g' = green.
# 'r' = red.
# 'c' = cyan.
# 'm' = magenta.
# 'y' = yellow.
# 'k' = black.
# 'w' = white.
```

```python

custom_size = 10
custom_marker = 'X'

x = np.arange(0,10,0.5)
y = x**2
z = (-1*x**2)+40

x1 = np.arange(2,12,0.5)
y1 = x**2
z1 = (-1*x**2)+70

plt.figure(figsize=(8,5))

plt.plot(x,y,custom_marker+' m',label = '$f(x)=x^2$', markersize = custom_size)
plt.plot(x,z,custom_marker+' y',label = '$f(x)=-x^2+20$', markersize = custom_size)
plt.plot(x,-y,custom_marker+' m',label = '$f(x)=x^2$', markersize = custom_size)
plt.plot(x,-z,custom_marker+' y',label = '$f(x)=-x^2+20$', markersize = custom_size)
plt.plot(x1,y1,custom_marker+' k',label = '$f(x)=x^2$', markersize = custom_size)
plt.plot(x1,z1,custom_marker+' k',label = '$f(x)=-x^2+20$', markersize = custom_size)
plt.plot(x1,-y1,custom_marker+' k',label = '$f(x)=x^2$', markersize = custom_size)
plt.plot(x1,-z1,custom_marker+' k',label = '$f(x)=-x^2+20$', markersize = custom_size)

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.suptitle('Plots for my functions', fontsize=30)

plt.subplot(2,2,3)
plt.plot(x,y,'o--b')
plt.title('$f(x)=x^2$')

plt.subplot(2,2,4)
plt.plot(x,z,'o:r')
plt.title('$f(x)=-x^2+20$')

plt.subplot(2,1,1)
plt.plot(x,y,'o--b',label = '$f(x)=x^2$')
plt.plot(x,z,'o:r',label = '$f(x)=-x^2+20$')
plt.title('$f(x)=-x^2+20$ and $f(x)=x^2$')
plt.tight_layout()
```

### Axes interface (object-oriented)

#### Matplotlib figures, axis and other parts of figure
https://matplotlib.org/stable/users/explain/quick_start.html#parts-of-a-figure

```python
x = np.arange(0,10,0.5)
y = x**2

fig = plt.figure(figsize = (8,5))
ax = fig.add_axes([0.1,0.1,0.9,0.9], facecolor="m")

ax.plot(x,y, color = 'black', lw = 2, ls='--', 
        marker = 'o', markersize = 8, markerfacecolor = 'w',
        markeredgecolor = 'w', markeredgewidth = 3)

ax.set_title('f(x)=$x^2$', family = 'serif',color = 'r',
         weight = 'bold',size = 16) 
ax.set_xlabel('x', fontdict = font)
ax.set_ylabel('y', fontdict = font)
ax.grid(color = 'k', ls = '--') 
# ax.set_facecolor('0.9')


plt.show()
fig.savefig('my_plot', dpi = 100, bbox_inches='tight')
```

```python
x = np.arange(0,10,0.5)
y = x**2
z = (-1*x**2)+20

plt.figure(figsize=(8,5))
plt.plot(x,y,'o--b',label = '$f(x)=x^2$')
plt.plot(x,z,'o:r',label = '$f(x)=-x^2+20$')
plt.legend()
plt.show()
```

```python
x = np.arange(0,10,0.5)
y = x**2
z = (-1*x**2)+20

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(x,y, 'o--b', label = '$f(x)=x^2$')
ax.plot(x,z, 'o:r', label = '$f(x)=-x^2+20$')
ax.legend(loc = 'best')
ax.set_title('Show two lines', 
             family = 'arial', color = '0.5',
             weight = 'bold',
             size = 16)
ax.set_xlabel('x', family = 'arial', size = 16)
ax.set_ylabel('y', family = 'arial', size = 16, rotation = 0)
ax.set_facecolor('0.9')

# add plot in plot
ax1 = fig.add_axes([0.65,0.4,0.2,0.2])
ax1.plot(x,y,x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim([2,4])
ax1.set_ylim([0,20])
ax1.text(2.7,15, '(3.2, 10)')
ax1.set_facecolor('0.7')
plt.show()

```

```python
fig_1 = plt.figure(figsize=(12,8), dpi=100, facecolor='gray') 
axes_1 = fig_1.add_axes([0,0,1,1]) #set xmin, ymin, width, and height 
axes_2 = fig_1.add_axes([0.2,0.2,0.5,0.5], facecolor='c',)
axes_3 = fig_1.add_axes([0.3,0.4,0.5,0.5], facecolor='green')
axes_2.plot(x,y, color = 'm')
axes_3.plot(x,y, color = 'k')

axes_3.patch.set_alpha(0.5)
```

```python
df_1 = df.sort_values('distance')

x = np.array(df_1.distance)
y = np.array(df_1.fare)
y1 = np.array(df_1.tip)

fig, (ax_dist_fare, ax_dist_tip) = plt.subplots(figsize=(10,6),
                                               ncols = 1,
                                               nrows = 2)
ax_dist_fare.plot(x,y, 'ro')
ax_dist_tip.plot(x,y1, 'ro')
```
