### pyplot interface (procedural, inheritance of matlab))

```python
x = np.arange(0,10,0.5)
y = x**2
plt.plot(x, y)
# plt.show()
```

```python
plt.plot(x,y,
         color='blue', #'b', '#1f77b4', (0,0,1)
         marker='o', 
         linestyle='dashed', # '--'
        )
```

```python
plt.plot(x,y,'o--b'); # fmt = '[marker][line][color]'
```

```python
# Add title, lables, limits
plt.plot(x,y,'o--b',markeredgecolor = 'r',)
plt.title('f(x)=$x^2$',
         family = 'serif',
         color = 'r',
         weight = 'bold',
         size = 16) 
plt.xlabel('x')
plt.ylabel('f(x)');
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y,'o--b',markeredgecolor = 'r', markersize = 7,
         label = '$f(x)=x^2$')
plt.title('f(x)=$x^2$',
         family = 'serif',
         color = 'g',
         weight = 'bold',
         size = 16) 
plt.xlabel('x', fontdict = font)
plt.ylabel('f(x)', fontdict = font)
plt.legend(loc='lower right')
plt.grid()
#plt.figure(figsize=(8,5));
```

```python
x = np.arange(0,10,0.5)
y = x**2
z = (-1*x**2)+20

plt.figure(figsize=(8,5))
plt.plot(x,y,'o--b',label = '$f(x)=x^2$')
plt.plot(x,z,'o:r',label = '$f(x)=-x^2+20$')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.suptitle('Plots for my functions', fontsize=30)

plt.subplot(2,2,3)
plt.plot(x,y,'o--b')
plt.title('$f(x)=x^2$')

plt.subplot(2,2,4)
plt.plot(x,z,'o:r')
plt.title('$f(x)=-x^2+20$')

plt.subplot(2,1,1)
plt.plot(x,y,'o--b',label = '$f(x)=x^2$')
plt.plot(x,z,'o:r',label = '$f(x)=-x^2+20$')
plt.title('$f(x)=-x^2+20$ and $f(x)=x^2$')
plt.tight_layout()
```

### Axes interface (object-oriented)

#### Matplotlib figures, axis and other parts of figure
https://matplotlib.org/stable/users/explain/quick_start.html#parts-of-a-figure

```python
x = np.arange(0,10,0.5)
y = x**2

fig = plt.figure(figsize = (8,5))
ax = fig.add_axes([0.1,0.1,0.9,0.9])

ax.plot(x,y, color = 'blue', lw = 2, ls='--',
        marker = 'o', markersize = 8, markerfacecolor = 'r',
        markeredgecolor = 'y', markeredgewidth = 3)

ax.set_title('f(x)=$x^2$', family = 'serif',color = 'r',
         weight = 'bold',size = 16) 
ax.set_xlabel('x', fontdict = font)
ax.set_ylabel('y', fontdict = font)
ax.grid(color = 'r', ls = '--') 
ax.set_facecolor('0.9')

plt.show()
fig.savefig('my_plot', dpi = 100, bbox_inches='tight')
```

```python
x = np.arange(0,10,0.5)
y = x**2
z = (-1*x**2)+20

plt.figure(figsize=(8,5))
plt.plot(x,y,'o--b',label = '$f(x)=x^2$')
plt.plot(x,z,'o:r',label = '$f(x)=-x^2+20$')
plt.legend()
plt.show()
```

```python
x = np.arange(0,10,0.5)
y = x**2
z = (-1*x**2)+20

fig, ax = plt.subplots(figsize = (8,5))
ax.plot(x,y, 'o--b', label = '$f(x)=x^2$')
ax.plot(x,z, 'o:r', label = '$f(x)=-x^2+20$')
ax.legend(loc = 'best')
ax.set_title('Show two lines', 
             family = 'arial', color = '0.5',
             weight = 'bold',
             size = 16)
ax.set_xlabel('x', family = 'arial', size = 16)
ax.set_ylabel('y', family = 'arial', size = 16, rotation = 0)
ax.set_facecolor('0.9')

# add plot in plot
ax1 = fig.add_axes([0.65,0.4,0.2,0.2])
ax1.plot(x,y,x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_xlim([2,4])
ax1.set_ylim([0,20])
ax1.text(2.7,15, '(3.2, 10)')
ax1.set_facecolor('0.7')
plt.show()

```

### Taxis dataset: calculate ride duration

```python
import pandas as pd
df_2 = df
df_2.head()
```

```python
df_2['ride_duration'] = (pd.to_datetime(df_2.dropoff) -  pd.to_datetime(df_2.pickup)).astype('timedelta64[m]')
print(df_2.ride_duration)
df_2.ride_duration.hist()
```

```python
df_2['fare_tip_ratio'] = round(df_2.tip/df_2.fare, 2)
print(df_2.fare_tip_ratio)
df_2.fare_tip_ratio.hist()
```

```python
df_2.head()
```

```python
df_2['fare_tip_ratio'].hist(bins=100,  
                            density=True, 
                            range=[0.001, 0.9],
                           figsize = (15,8))
```
